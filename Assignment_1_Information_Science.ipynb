```python
import csv

# Search for top 50 videos
search_response = youtube.search().list(
    q="avatar movie",
    type="video",
    part="id",
    maxResults=50,
    regionCode="US",
    order="relevance"
).execute()

# Extract video IDs
video_ids = [item['id']['videoId'] for item in search_response['items']]

# Create a CSV file and write the data
with open('avatar_movie_stats.csv', 'w', newline='') as csvfile:
    fieldnames = ['Video ID', 'Title', 'Views', 'Likes', 'Comments']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    writer.writeheader()

    for video_id in video_ids:
        video_response = youtube.videos().list(
            id=video_id,
            part="statistics"
        ).execute()

        video_stats = video_response['items'][0]['statistics']

        writer.writerow({
            'Video ID': video_id,
            'Title': video_title,
            'Likes': video_stats['likeCount'],
            'Comments': video_stats['commentCount']
        })
```

```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('avatar_movie_stats.csv')

# Sort by comments in descending order
top_10_comments = df.sort_values(by='Comments', ascending=False).head(10)

# Extract video IDs and titles
top_10_video_ids = top_10_comments['Video ID'].tolist()
top_10_video_titles = top_10_comments['Title'].tolist()

print("Top 10 Videos with Highest Comments:")
print(top_10_comments[['Video ID', 'Title']])
```

```python
def get_video_comments(video_id):
    comments = []
    next_page_token = None  # Initialize to None

    while True:
        results = youtube.commentThreads().list(
            part="snippet",
            videoId=video_id,
            textFormat="plainText",
            maxResults=100,
            pageToken=next_page_token  # Use the next_page_token for pagination
        ).execute()

        for item in results['items']:
            comment = item['snippet']['topLevelComment']['snippet']['textDisplay']
            comments.append(comment)

        # Check if there are more pages of comments
        next_page_token = results.get('nextPageToken')

        if not next_page_token:
            break  # No more comments to retrieve

    return comments

# Retrieve comments for the top 10 videos
for video_id, video_title in zip(top_10_video_ids, top_10_video_titles):
    comments = get_video_comments(video_id)
    print(f"Video Title: {video_title}")
    print("Comments:")
    for comment in comments:
        print(comment)
```

```python
import json
from googleapiclient.errors import HttpError

# Create a dictionary to store the comments for each video
video_comments_dict = {}

# Retrieve comments for the top 10 videos
for video_id, video_title in zip(top_10_video_ids, top_10_video_titles):
    try:
        comments = get_video_comments(video_id)
        video_comments_dict[video_title] = comments
    except HttpError as e:
        # Handle the error (e.g., print an error message or skip the video)
        print(f"Error for Video ID: {video_id}")
        print(e)

# Export the dictionary to a JSON file
with open('top_10_comments.json', 'w') as json_file:
    json.dump(video_comments_dict, json_file)
```

```python
def calculate_likes_vs_views_ratio(likes, views):
    return int(likes) / int(views)

# Calculate and print the likes vs. views ratio for the top 10 videos
for video_id, video_title in zip(top_10_video_ids, top_10_video_titles):
    video_response = youtube.videos().list(
        id=video_id,
        part="statistics"
    ).execute()

    video_stats = video_response['items'][0]['statistics']
    likes = video_stats['likeCount']
    views = video_stats['viewCount']

    ratio = calculate_likes_vs_views_ratio(likes, views)

    print(f"Video Title: {video_title}")
    print(f"Likes: {likes}")
    print(f"Views: {views}")
    print(f"Likes vs. Views Ratio: {ratio:.2%}")
    print()
```
